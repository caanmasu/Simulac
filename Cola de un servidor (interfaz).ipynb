# Simulación de una cola de un servidor 

```python
from tkinter import *
from tkinter import ttk
from random import *
import time
```

```python
t = tsuc = Nll = Ns = n = 0
M = 2**31
TSuctll = M
TSucts = M
tamano = 1000
LL = [0]*tamano
S = [0]*tamano
Serv = [0]*tamano


def Iniciar(T, atendidos, cola, ventana, label_med_sistema, label_med_cliente, retardo, VSimulacion):
    global TSuctll, TSucts, tsuc, M, t, n, Ns
    reiniciarValores(t,tsuc,Nll,Ns,n,TSuctll,TSucts,LL,S,Serv,label_med_sistema,label_med_cliente, atendidos, ventana)
    cargarSimulacion(ventana, VSimulacion)    
    
    X = GCM()

    if (X>T):
        Tp = t_med_sistema = t_med_cola = 0
    else:
        Llegada(X, ventana, cola, T, VSimulacion, retardo)
        while (TSuctll!=M or TSucts!=M):
            if (TSuctll<TSucts):
                tsuc = TSuctll
                TSuctll = M
                Llegada(tsuc, ventana, cola, T, VSimulacion, retardo)
            if (TSucts<TSuctll):
                tsuc = TSucts
                TSucts = M
                Servidor(tsuc, ventana, cola, VSimulacion, atendidos, retardo)                
        Tp = max(0, t-T)
        acumulo1 = acumulo2 = ind = 0
        while (ind<Nll):
            acumulo1 = acumulo1 + S[ind] - LL[ind]
            acumulo2 = acumulo2 + S[ind] - LL[ind] - Serv[ind]
            ind = ind + 1
        t_med_sistema = acumulo1/Nll
        t_med_cola = acumulo2/Nll
        label_med_sistema.set(t_med_sistema)
        label_med_cliente.set(t_med_cola)
        ventana.update()
        print ("Tiempo:",T)
        print ("Clientes en espera:",n)
        print ("Clientes atendidos:",Ns)
        print ("Tiempo medio del sistema en atender un cliente: ", t_med_sistema)
        print ("Tiempo promedio que gasta un cliente en cola:",t_med_cola)
        
def reiniciarValores(t,tsuc,Nll,Ns,n,TSuctll,TSucts,LL,S,Serv,label_med_sistema,label_med_cliente, atendidos, ventana):
    t = tsuc = Nll = Ns = n = 0
    TSuctll = M
    TSucts = M
    tamano = 1000
    LL = [0]*tamano
    S = [0]*tamano
    Serv = [0]*tamano
    label_med_sistema.set("")
    label_med_cliente.set("")
    atendidos.set("")
    ventana.update()
        
def cargarSimulacion(ventana, VSimulacion):
    for i in range(20):
        LLlenar = Label(VSimulacion, text=" ").grid(row=1, column=i)
    for j in range(23,33):
        LLlenar = Label(VSimulacion, text=" ").grid(row=1, column=j)
    LServidor = Label(VSimulacion, text="▓").grid(row=1, column=20)
    LServidor = Label(VSimulacion, text="▓").grid(row=1, column=21)
    LServidor = Label(VSimulacion, text="▓").grid(row=1, column=22)
    for i in range(1, 20):
            LCliente = Label(VSimulacion, text="웃").grid(row=1, column=i) #1, 22
            LCliente = Label(VSimulacion, text=" ").grid(row=1, column=i) #1, 22
            ventana.update()
    for j in range(23, 40):
            LCliente = Label(VSimulacion, text="웃").grid(row=1, column=j) #1, 22
            LCliente = Label(VSimulacion, text=" ").grid(row=1, column=j) #1, 22
            ventana.update()
    
def Llegada(tsuc_local, ventana, cola, T, VSimulacion, retardo):
    global tsuc, n, Nll, t, TSucts, TSuctll, Ns
    tsuc = tsuc_local
    t = tsuc
    n = n + 1
    cola.set(n)
    clienteCola(VSimulacion, ventana, n, retardo)
    ventana.update()
    Nll = Nll + 1
    LL[Nll] = t
    X = GCM()
    if ((t+X)<T):
        TSuctll = t + X
    if (n == 1):
        Y = GCM()
        TSucts = t + Y
        Serv[Ns + 1] = Y
        
def clienteCola(VSimulacion, ventana, n, retardo):
    bmax = 21
    j = 1
    for i in range(1,(bmax-n)):
        if (n<j):
            LCliente = Label(VSimulacion, text="웃").grid(row=1, column=i) #1, 22
            ventana.update()
        if (i>1):
            LCliente = Label(VSimulacion, text=" ").grid(row=1, column=(i-1)) #1, 22
            ventana.update()
        time.sleep(retardo.get())
        j = j + 1
        
def Servidor(tsuc_local, ventana, cola, VSimulacion, atendidos, retardo):
    global tsuc, t, n, Ns, TSucts
    tsuc = tsuc_local
    t = tsuc
    n = n - 1
    Ns = Ns + 1
    cola.set(n)
    atendidos.set(Ns)
    clienteAtendido(ventana, VSimulacion, retardo)    
    ventana.update()
    S[Ns] = t
    if (n>0):
        Y = GCM()
        TSucts = t + Y
        Serv[Ns] = Y
        
def clienteAtendido(ventana, VSimulacion, retardo):
    bmax = 21
    LCliente = Label(VSimulacion, text=" ").grid(row=1, column=19)
    for j in range(1,20):
        if (n==j):
            for k in range(n):
                LCliente = Label(VSimulacion, text="웃").grid(row=1, column=20-(k+1))
                ventana.update()
                time.sleep(retardo.get())
    LCliente = Label(VSimulacion, text=" ").grid(row=1, column=19-n)
    for i in range(1,10):
        LCliente = Label(VSimulacion, text="웃").grid(row=1, column=bmax+2+(i-1)) #1, 22
        ventana.update()
        if (i>1):
            LCliente = Label(VSimulacion, text=" ").grid(row=1, column=bmax+1+(i-1)) #1, 22
            ventana.update()
            time.sleep(retardo.get())
    LCliente = Label(VSimulacion, text=" ").grid(row=1, column=bmax+10)

        
def GCM():
    m = 2**32
    X0 = randint(1, 90)
    X = (25214903917*X0+1)%m
    return X/m

def Interfaz():
    v0 = Tk()
    v0.title("Simulación de colas de un servidor")
    #v0.config(bg="red")
    LabelTitulo = Label(v0, text="Simulación de colas de un servidor").pack()
    LabelIngreso = Label(v0, text="Ingrese tiempo").pack()
    Tiempo = IntVar()
    EntradaTiempo = Entry(v0, textvar=Tiempo).pack()
    LabelRetardo = Label(v0, text="Ingrese retardo (microsegundos)").pack()
    Retardo = DoubleVar()
    RetardoTiempo = Entry(v0, textvar=Retardo).pack()
    VarClientes = StringVar()
    InfoClientes = Label(v0, text="Clientes atendidos:").pack()
    ClientesVar = Label(v0, textvar=VarClientes).pack()   
    VarEnCola = StringVar()
    InfoEspera = Label(v0, text="Clientes en espera:").pack()
    EsperaVar = Label(v0, textvar=VarEnCola).pack()
    
    InfoPromedio1 = Label(v0, text="Tiempo medio del sistema:").pack()
    VarSistema = StringVar()
    SistemaVar = Label(v0, textvar=VarSistema).pack()
    
    InfoPromedio2 = Label(v0, text="Tiempo medio cliente:").pack()
    VarMedioCliente = StringVar()
    PromedioCliente = Label(v0, textvar=VarMedioCliente).pack()
    VSimulacion = Toplevel(v0)
    VSimulacion.geometry("500x30")
    VSimulacion.title("Gráfico de simulación")
    BIniciar = Button(v0, text="Iniciar", command=lambda: Iniciar(Tiempo.get(), VarClientes, VarEnCola, v0, VarSistema, VarMedioCliente, Retardo, VSimulacion)).pack()
    
    
    
    
    
    v0.mainloop()
        
if __name__ == "__main__":
    Interfaz()
```
