# Juego de Space War

```python
from tkinter import *
import time
from random import randint
import threading
```

```python
j = 13
p = 19
q = 1
x = 10
y = 5
MetDestruidos = 0
flag = True
flag2 = True
flag3 = True
flagTiempo = True
VPrincipal = Tk()
VEstad = Toplevel(VPrincipal)
Lista = Listbox(VEstad)
nivel = IntVar()
TReal = 0

    
def Estadisticas():
    global VPrincipal, MetDestruidos, VEstad
    Lista.pack()
    Lista.insert(1, "Meteoros destruidos: ")
    Lista.insert(2, MetDestruidos)
    Lista.insert(3, "Tiempo transcurrido: ")
    Lista.insert(4, "0")
    VEstad.title("Estadísticas")
    VEstad.update()
    
def crearHilos():
    global nivel, flag
    flag = True
    n = nivel.get()
    if (nivel.get()==0): n = nivel.get()+1
    for i in range(n):
        threads=list()
        t = threading.Thread(target=lanzarMeteoro)
        threads.append(t)
        t.start()

    
def cargarTiempo():
    threads2=list()
    time = threading.Thread(target=actualizarTiempo)
    threads2.append(time)
    time.start()
    
    
def cargarPantalla():
    global VPrincipal
    VPrincipal.title ("Space War")
    VPrincipal.geometry("500x500")
    for i in range(22): #26
        for j in range(43): #43
            LLlenar = Label(VPrincipal, text="cc").grid(row=i, column=j)
            LLlenar = Label(VPrincipal, text="  ").grid(row=i, column=j)
            VPrincipal.update()
            
def crearBarra():
    global VPrincipal, BDisparo
    BarIzq = Label(VPrincipal, text="█").grid(row=20, column=12)
    BarCen = Label(VPrincipal, text="█").grid(row=20, column=13)
    BarDer = Label(VPrincipal, text="█").grid(row=20, column=14)
    VPrincipal.update()
    
def crearBotones():
    global VPrincipal, VEstad, LNivel, ENivel
    BIzq = Button (VPrincipal, text="<", command=lambda: correrBarraIzq()).grid(row=22, column=12)
    BDer = Button (VPrincipal, text=">", command=lambda: correrBarraDer()).grid(row=22, column=14)
    BReinicio = Button(VPrincipal, text="↻", command=lambda: reiniciarTodo()).grid(row=22, column=17)
    BDisparo = Button (VPrincipal, text="│",command=lambda: Disparar()).grid(row=22,column=13)
    LNivel = Label(VEstad, text="Añadir meteoros: ").pack()
    ENivel = Entry(VEstad, textvar=nivel).pack()
    
    
def reiniciarTodo():
    global j, p, q, MetDestruidos, flag, nivel
    MetDestruidos = 0
    j = 13
    p = 19
    q = 1
    flag = False
    TReal = 0
    cargarPantalla()
    crearBarra()
    Lista.delete(0,END)
    Estadisticas()
    time.sleep(4)
    for i in range(nivel.get()):
        crearHilos()
    flag = True
    lanzarMeteoro()
    
def correrBarraIzq():
    global j, VPrincipal
    for i in range(1,3):
        BarDes = Label(VPrincipal, text="█").grid(row=20, column=j-i)
    BarRem = Label(VPrincipal, text="  ").grid(row=20, column=j+1)
    j = j - 1
    VPrincipal.update()
    
def correrBarraDer():
    global j, VPrincipal
    for i in range(1,3):
        BarDes = Label(VPrincipal, text="█").grid(row=20, column=j+i)
    BarRem = Label(VPrincipal, text="  ").grid(row=20, column=j-1)
    j = j + 1
    VPrincipal.update()
    
def Disparar():
    global j, p, q, x, VPrincipal, MetDestruidos, flag2, flag3
    u = j
    if (flag3==True):
        for i in range(20):
            flag3=False
            x = u
            LDisparo = Label(VPrincipal, text="│").grid(row=19-i, column=u)
            time.sleep(0.03)
            p = 19-i
            VPrincipal.update()
            LDisparo = Label(VPrincipal, text="  ").grid(row=19-i, column=u)
            if (p==q):
                if (x==y):
                    flag2 = False
                    MetDestruidos = MetDestruidos + 1
                    Lista.insert(2, MetDestruidos)
                    Lista.delete(1, None)
                    VEstad.update()
                    break
        flag3=True
        
def lanzarMeteoro():
    global q, p, x, y, VPrincipal, flag, flag2, flagTiempo
    while (flag):
        r = randint(1,26)
        flag2 = True
        for i in range(20):
            if (flag2==False):
                break
            y = r
            LMeteoro = Label(VPrincipal, text="☠").grid(row=1+i, column=r)
            time.sleep(0.2)
            if (i==19):
                if (j==r or (j-1)==r or (j+1)==r):
                    flag = False
                    flagTiempo = False
                    gameOver()
                    break
            q = 1+i
            VPrincipal.update()
            LMeteoro = Label(VPrincipal, text="  ").grid(row=1+i, column=r)
        
def actualizarTiempo():
    global TReal, flagTiempo
    TReal = 0
    TInicio = time.time()
    while (flagTiempo):
        time.sleep(1)
        TReal = time.time()-TInicio
        Lista.insert(4, int(TReal))
        Lista.delete(3, None)
        VEstad.update()
        
def gameOver():
    global VPrincipal
    LGameOver = Label(VPrincipal, text="G").grid(row=14, column=10)
    LGameOver = Label(VPrincipal, text="a").grid(row=14, column=11)
    LGameOver = Label(VPrincipal, text="m").grid(row=14, column=12)
    LGameOver = Label(VPrincipal, text="e").grid(row=14, column=13)
    LGameOver = Label(VPrincipal, text="O").grid(row=16, column=10)
    LGameOver = Label(VPrincipal, text="v").grid(row=16, column=11)
    LGameOver = Label(VPrincipal, text="e").grid(row=16, column=12)
    LGameOver = Label(VPrincipal, text="r").grid(row=16, column=13)
    Lista.insert(END, "Has perdido")


if __name__=="__main__":
    cargarPantalla()
    Estadisticas()
    crearBarra()
    crearBotones()
    cargarTiempo()
    crearHilos()
    VPrincipal.mainloop()
```
