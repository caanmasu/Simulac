```python
from pylab import *
import matplotlib.pyplot as plt
import numpy as np
from random import random
import math

suma = 0
n = 0
x = 0
y = 0

print ("Integral por el método de Monte Carlo \n\nI = ∫e^(-x^2)dx - (0, 2)\n")

Y = []
X = []

while n <= 1000000:
    x = 2*random() #X - Unif(0,2)
    y = 2*math.exp(-x*x) #
    suma = suma+y 
    
    if (n==10000):
        print("n = 10000 -> ", suma/n) #suma/n es el promedio de y para cada valor de x
    if (n==100000):
        print("n = 100000 -> ", suma/n)
    if (n==1000000):
        print("n = 1000000 -> ", suma/n)
    
#//////Vectores para la gráfica///////
    Y.append(suma/(n+1))
    X.append(n)
#////////////////////////////////////
        
    n = n + 1

print ("\n\nGráficas de 10000, 100000 y 1000000 iteraciones respectivamente")

#////////////Gráficas////////////////////////////
subplot(131)
plt.xlabel("Iteraciones")
plt.ylabel("∫e^(-x^2)dx")
plt.grid(True)
plt.plot(X[1:10000], Y[1:10000], '-y')
subplot(132)
plt.xlabel("Iteraciones")
plt.grid(True)
plt.plot(X[1:100000], Y[1:100000], '-b')
subplot(133)
plt.xlabel("Iteraciones")
plt.grid(True)
plt.plot(X, Y, 'r')
plt.show()
#/////////////////////////////////////////////////
```

```python
from pylab import *
import matplotlib.pyplot as plt
import numpy as np
import math

#Método numérico de integración Regla del Trapecio compuesto

a = 0 #Límite inferior
b = 2 #Límite superior
n = 10000 #Corresponde a la cantidad de subintervalos
h = (b-a)/n #Ancho del subintervalo
Fa = math.exp(-a*a)
Fb = math.exp(-b*b)

print ("Integral por el método de Trapecio compuesto \n\nI = ∫e^(-x^2)dx - (0, 2)\n")

X = []
Y = []
s = 0
for k in range(1, n):
    s=s+math.exp(-(a+k*h)*(a+k*h))    
    X.append(k)
    Y.append(h*((Fa+Fb)/2+s))
    
resultado = h*((Fa+Fb)/2+s) #Fórmula final, donde la altura del trapecio es (Fa+Fb)/2+s

print ("I = ", resultado, "para", n, "iteraciones")

plt.title("Gráfica para 10000 iteraciones")
plt.xlabel("Iteraciones")
plt.ylabel("∫e^(-x^2)dx")
plt.grid(True)
plt.plot(X[1:10000], Y[1:10000], '-g')
plt.show()
```
